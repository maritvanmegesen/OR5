```python
import pandas as pd

#implementing constructive heuristic for line production
#description to be added
df_ch=pd.read_excel('Line Production September 2023.xlsx')
df_ch.head(27)
```

```python
#sorting products based on 
df_ch['proportion']=df_ch['deadline']/df_ch['penalty cost']
df_ch.head()
```

```python
#sort the rows by urgency/by deadline - could also do by penalty cost
df_ch=df_ch.sort_values(by='proportion')
df_ch=df_ch.reset_index(drop=True)
df_ch.head()
```

```python
df_ch['ideal line - ch']=''
line_columns=['L1', 'L2', 'L3', 'L4', 'L5', 'L6', 'L7']
for product in df_ch.index:
    ideal_line=df_ch[line_columns].idxmin(axis=1) #chooses the index of the column instead of processing time
    df_ch['ideal line - ch']=ideal_line
df_ch.head(20)
```

```python
#of course we will not always use the ideal column so this is much more helpful
def sort_column_names(row):
    sorted_columns = sorted(line_columns, key=lambda col:row[col])
    return ', '.join(sorted_columns)
df_ch['succesion of best lines - ch']=df_ch.apply(sort_column_names, axis=1)
df_ch.head(20)
```
